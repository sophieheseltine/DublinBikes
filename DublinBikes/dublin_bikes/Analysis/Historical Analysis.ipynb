```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SELECT_t___FROM_DublinBikes_UserTrends_t.csv')
df.head()
```

```python
df.dtypes
```

```python
#separating dates from time
df1 = pd.DataFrame(df.last_update.str.split(' ',1).tolist(), columns = ['date','time'])
df1.head()
df.head()
```

```python
df = df[['name', 'latitude', 'longitude', 'available_bikes', 'available_bike_stands', 'bike_stands', 'status']]
df.head()
```

```python
frames = [df1, df]
df = pd.concat(frames, axis=1)
df.head()
```

```python
df['day'] = np.where(df['date']=='29/03/2018', 'Thursday', 'no')
```

```python
HourList = []
for row in df.iterrows():
   split = row[1]['time'].split(':')
   HourList.append(split[0])
```

```python
df_Hour = pd.DataFrame({'hour':HourList})
df_Hour['hour'] = df_Hour['hour'].astype('int64')
df_new = pd.concat([df, df_Hour], axis=1)
df_new.head()
```

```python
dfGrouped = df_new[['hour', 'name', 'bike_stands', 'available_bikes', 'available_bike_stands', 'day', 'status']].groupby(['name', 'day', 'hour']).mean()
dfGrouped.head(30)
```

```python
# Write the dataframe with averages and groupbys into a new csv called dfGroupedFinal.csv
dfGrouped.add_suffix('_mean').reset_index().to_csv('dfGroupedFinal.csv', index=False)
```

```python
dfWeatherG = df_new[['date', 'hour', 'name', 'bike_stands', 'available_bikes', 'available_bike_stands', 'status']].groupby(['name', 'date', 'hour']).mean()
dfWeatherG.head(30)
```

```python
dfWeatherG.add_suffix('_mean').reset_index().to_csv('dfWeatherG.csv', index=False)
```
