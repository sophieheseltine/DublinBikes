```python
dfWeather = dfWeather.rename(columns={'Hour':'hour', 'Day':'date'})
dfWeather.head()
```

```python
dfStations = dfStations[['name', 'bike_stands_mean', 'available_bikes_mean', 'available_bike_stands_mean', 'hour', 'date']]
dfStations.head()
```

```python
# Merge stations and weather based on hour and date
dfSW = pd.merge(dfStations, dfWeather, on=['hour', 'date'], how='outer')
dfSW
```

```python
dfSW.to_csv('dfSW.csv', index=False)
```

# Regression

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
dfSW['rainbin'] = number.fit_transform(dfSW['Rain'].astype('int'))
```

```python
dfSW.dtypes
```

## Linear Regression

```python
import statsmodels.formula.api as sm

lm = sm.ols(formula="available_bikes_mean ~  rainbin + hour + C(name)", data=dfSW).fit()
lm.summary()
```

```python
import matplotlib.pyplot as plt

plt.figure()
flierprops = dict(marker='o', markerfacecolor='green', markersize=6,
                  linestyle='none')
bp = dfSW.boxplot(column=['available_bikes_mean'], by=['rainbin'], flierprops=flierprops, figsize=(10, 7))
plt.show()
```

From the above box plots, we can see that on the whole dataset, there is virtually no difference between the availability of bikes based on rain.

```python
# Decided to add the days back in

dfSW['day'] = pd.np.where(dfSW.date.str.contains("29/03/2018"), "Thursday",
                   pd.np.where(dfSW.date.str.contains("30/03/2018"), "Friday",
                   pd.np.where(dfSW.date.str.contains("31/03/2018"), "Saturday",
                    pd.np.where(dfSW.date.str.contains("01/04/2018"), "Sunday", 
                    pd.np.where(dfSW.date.str.contains("02/04/2018"), "Monday",
                   pd.np.where(dfSW.date.str.contains("03/04/2018"), "Tuesday",
                    pd.np.where(dfSW.date.str.contains("04/04/2018"), "Wednesday",
                    pd.np.where(dfSW.date.str.contains("05/04/2018"), "Thursday",
                   pd.np.where(dfSW.date.str.contains("06/04/2018"), "Friday",
                    pd.np.where(dfSW.date.str.contains("07/04/2018"), "Saturday",
                    pd.np.where(dfSW.date.str.contains("08/04/2018"), "Sunday",
                   pd.np.where(dfSW.date.str.contains("09/04/2018"), "Monday",
                    pd.np.where(dfSW.date.str.contains("10/04/2018"), "Tuesday",
                    pd.np.where(dfSW.date.str.contains("11/04/2018"), "Wednesday",
                    pd.np.where(dfSW.date.str.contains("12/04/2018"), "Thursday", "null"      
                               
                               )))))))))))))))
dfSW.head()
```

```python
dfSW['daytype'] = pd.np.where(dfSW.day.str.contains("Monday"), "Weekday",
                   pd.np.where(dfSW.day.str.contains("Tuesday"), "Weekday",
                   pd.np.where(dfSW.day.str.contains("Wednesday"), "Weekday",
                    pd.np.where(dfSW.day.str.contains("Thursday"), "Weekday", 
                    pd.np.where(dfSW.day.str.contains("Friday"), "Weekday",
                   pd.np.where(dfSW.day.str.contains("Saturday"), "Weekend",
                    pd.np.where(dfSW.day.str.contains("Sunday"), "Weekend", "no"                               
                               )))))))
dfSW.head()
```

```python
sns.factorplot(x="hour", y="available_bikes_mean", hue="rainbin", col="daytype", data=dfSW, kind="bar", size=16, palette="Blues_d")
plt.show()
```

```python
#dealing with categorical datatypes

name_dummies = pd.get_dummies(dfSW.name, prefix='Station')

daytype_dummies = pd.get_dummies(dfSW.daytype, prefix='daytype')
dfSW_reg = pd.concat([dfSW, daytype_dummies, name_dummies], axis=1)
```

```python
dfSW_reg.head()
```

```python
dfSW_reg = dfSW_reg.drop(['name', 'date', 'day', 'daytype', 'Rain', 'available_bike_stands_mean', 'bike_stands_mean'], axis=1)
```

```python
feature_list1 = list(dfSW.columns)
feature_list = list(dfSW_reg.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

df_train, df_validation = train_test_split(dfSW_reg[dfSW_reg.available_bikes_mean.notnull()], test_size = 0.3)
df_learn = dfSW_reg[dfSW_reg.available_bikes_mean.notnull()]

#feature_columns = dfSW_reg.select_dtypes(['uint8', 'int64']).columns

rf2 = RandomForestRegressor(oob_score=True)
dependent_variable = ['available_bikes_mean']

x = dfSW_reg[dfSW_reg.columns.difference(dependent_variable)]
y = dfSW_reg[dependent_variable]

rf2.fit(x, y)

#df_validation = df_validation.assign(rf_bikes = rf2.predict(df_validation[['hour', 'daytype_Weekday', 'daytype_Weekend', 'rainbin']]))
#mean_absolute_error(df_validation['available_bikes_mean'], df_validation['rf_bikes'], sample_weight=None, multioutput='uniform_average')
                    
print("OOB accuracy: ")
print(rf2.oob_score_)
```
